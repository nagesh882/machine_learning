```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/ubuntu/Desktop/machine_learning/linear-regression/canada_per_capita_income.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
df.columns = df.columns.str.strip()
plt.scatter(df["year"], df["per capita income (US$)"], color="red", marker="*", label="Data Points")
plt.plot(df["year"], df["per capita income (US$)"], color="blue", label="Regression Line")
plt.legend()
plt.show()
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(df[["year"]], df["per capita income (US$)"])
```

```python
prediction = model.predict(pd.DataFrame({'year': [2024]}))
print("Prediction of canada per capita income (US$):", prediction[0])
```
